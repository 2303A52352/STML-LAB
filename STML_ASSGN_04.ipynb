```python
import pandas as pd
import zipfile

# Open the ZIP file
with zipfile.ZipFile("/content/archive (6).zip") as z:
  # Get a list of files in the archive
  file_list = z.namelist()
  # Print the list of files to see which one you want to read
  print(file_list)

  # You mentioned 'price_range' might be in 'train.csv', so let's try reading that
  with z.open('train.csv') as f: # Changed from 'test.csv' to 'train.csv'
    # Read the CSV file into a pandas DataFrame
    d = pd.read_csv(f)
    print(d)

    # Now try accessing 'price_range' within the scope of d
    y = d['price_range']
    print(y)

  # if you want to read test.csv replace train.csv with test.csv in the line above
```

```python
x=d.drop('price_range',axis=1)
print(x)
```

```python
dd=(x-x.min())/(x.max()-x.min())
print(dd)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dd,y,test_size=10)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
import pandas as pd
import zipfile

# Open the zip file
with zipfile.ZipFile("/content/archive (7).zip") as z:
    # Get a list of files in the zip
    file_list = z.namelist()
    # Assuming you want to read 'train.csv', change this if needed
    csv_file = 'train.csv'
    # Check if the desired file is in the zip
    if csv_file in file_list:
        # Open the CSV file within the zip
        with z.open(csv_file) as f:
            # Read the CSV data into a pandas DataFrame
            d = pd.read_csv(f)
    else:
        print(f"Error: {csv_file} not found in the zip file.")

print(d)
```

```python
print(d.dtypes)
```

```python

print(d.shape)
```

```python
target=d['price_range']
print(target)
```

```python
feature = d.drop('price_range',axis=1)
print(feature)
```

```python
d.isnull()
```

```python
normalised_data=(feature-feature.min())/(feature.max()-feature.min())
print(normalised_data)
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(normalised_data,target,test_size=100)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
